```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from mlxtend.data import iris_data
from mlxtend.plotting import plot_decision_regions

# Initializing Classifiers
labels = ['Logistic Regression', 'Random Forest', 'RBF kernel SVM']
clf = []
clf.append(LogisticRegression(random_state=0))
clf.append(RandomForestClassifier(random_state=0))
clf.append(SVC(random_state=0, probability=True))

# Loading some example data
X, y = iris_data()
X = X[:,[0, 2]]

# Plot classifiers
for i in range(len(clf)):
    clf[i].fit(X, y)
    plt.figure(figsize=(20,7))
    fig = plot_decision_regions(X=X, y=y, clf=clf[i], legend=0)
    plt.title(labels[i], fontsize=20)
    plt.show()
```
