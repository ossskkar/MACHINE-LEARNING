```python
# PROJECT: Use Central Station dataset (cvpr2015) to implement K-means clustering

# Dataset description
#
# Column 1: Timestamp
# Column 2: Track id
# Column 3: x position
# Column 4: y position
```

```python
# Import Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from mpl_toolkits.mplot3d import Axes3D
import sklearn.metrics as sm

from sklearn.cluster import KMeans

import matplotlib.image as mpimg
```

```python
# Load Data
dataset = pd.read_csv("../data/cstation_cvpr2015.csv", header=None)
X = dataset.iloc[:,2:4]
```

```python
# Create K-Means Model
c = 10 # number of clusters
rs = 5 # number of random states

kmeans = KMeans(n_clusters = c, random_state=rs)
kmeans.fit(X)
```

```python
# Plot Results
plt.figure
plt.subplots(1,3,figsize=(50,10))

# Illustration of the environment
plt.subplot(1,3,1)
img=mpimg.imread('../images/cstation.jpg')
plt.imshow(img)

# Data samples
plt.subplot(1,3,2)
plt.scatter(x=X.iloc[:,0], y=X.iloc[:,1],s=1,c='green')

# Clusters
plt.subplot(1,3,3)
for i in range(c) :
    plt.scatter(x=X.iloc[kmeans.labels_ == i,0], y=X.iloc[kmeans.labels_ == i,1],s=1)
    plt.scatter(x=kmeans.cluster_centers_[i,0], y=kmeans.cluster_centers_[i,1],c='black',marker="P",s=200)

plt.show()

print('Number of samples ->', len(X))
print('Number of clusters ->', c)
```
